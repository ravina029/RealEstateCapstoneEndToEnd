```python
#wordcloud_df = df1.merge(df, left_index=True, right_index=True)
#[['features','sector']]
wordcloud_df = df1.merge(df, left_index=True, right_index=True)[['sector','features']]
```

```python
wordcloud_df.head()
```

```python
import ast
main = []
for item in wordcloud_df['features'].dropna().apply(ast.literal_eval):
    main.extend(item)
```

```python
main
```

```python
from wordcloud import WordCloud
```

```python
feature_text = ' '.join(main)
```

```python
#feature_text
```

```python
import pickle
pickle.dump(feature_text, open('feature_text.pkl','wb'))
```

```python
feature_text
```

```python
plt.rcParams["font.family"] = "Arial"

wordcloud = WordCloud(width = 800, height = 800, 
                      background_color ='white', 
                      stopwords = set(['s']),  # Any stopwords you'd like to exclude
                      min_font_size = 10).generate(feature_text)

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show() # st.pyplot()
```

```python
def transform_feature(feature):
    # Perform your transformation here
    transformed_feature = feature.upper()  # Example transformation: Convert to uppercase
    return transformed_feature

# Apply transformation to each entry in the feature column
wordcloud_df['features'] = wordcloud_df['features'].apply(lambda x: [transform_feature(item) for item in ast.literal_eval(x)] if pd.notnull(x) else x)
```
